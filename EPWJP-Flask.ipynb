````python
from flask import Flask
````

````python
app = Flask(__name__)
````

````python
@app.route('/hello', methods=['GET'])
def get_hello():
    return 'GET hello'
````

### In real world:
```python
if __name__ == '__main__':
    app.run(debug=True)
```

### Jupyter notebook:

````python
from werkzeug.serving import run_simple
if __name__ == '__main__':
    run_simple('localhost', 9000, app)
````

### Other HTTP methods:

````python
@app.route('/hello2', methods=['GET'])
def get_hello2():
    return 'GET hello2'

@app.route('/hello', methods=['POST'])
def post_hello():
    return 'POST hello'

@app.route('/hello', methods=['PUT'])
def put_hello():
    return 'PUT hello'

@app.route('/hello', methods=['DELETE'])
def delete_hello():
    return 'DELETE hello'
````
